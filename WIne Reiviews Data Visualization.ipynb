```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from wordcloud import WordCloud, STOPWORDS 
```

```python
sns.set_style('darkgrid')
wordcloud = WordCloud(width = 1000, height = 400, 
                background_color ='white',  
                min_font_size = 10).generate(str(wine.description)) 
# Generate plot of most popular words in wine descriptions
plt.figure(figsize = (10, 10), facecolor = None) 
plt.tight_layout(pad = 0) 
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
sns.set_style('darkgrid')
wordcloud = WordCloud(width = 1000, height = 400, 
                background_color ='white',  
                min_font_size = 10).generate(str(wine.variety)) 
# Generate plot of most popular words in country
plt.figure(figsize = (10, 10), facecolor = None) 
plt.tight_layout(pad = 0) 
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
